```python
import yfinance as yf
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import plotly_express as px
```

uso pandas para traer desde wikipedia una lista con los symbolos de las acciones

```python
dfs = pd.read_html('https://en.wikipedia.org/wiki/List_of_S%26P_500_companies')[0]
dfs
```

guardo en una lista los symbol

```python
dfs_symbol = list(dfs['Symbol'])
dfs_symbol[65] = 'BRK-B'
dfs_symbol[81] = 'BF-B'
dfs_symbol
```

desde la api yfinance de yahoo traigo todas las companias que componen el sp500

```python
sp500_1 = yf.download(    tickers=dfs_symbol, start='2000-01-01', end='2021-12-31', 
                        interval='1d', group_by='ticker') 
```

```python
sp500_1.sort_index(inplace=True)
```

```python
sp500_1_copy = sp500_1.copy()
```

#### Creacion de los dataframe

```python
#crea un dataframe para guardar unicamente la columna Adj Close
data_adj_close=pd.DataFrame()
for i in dfs_symbol:
    data_adj_close= pd.concat([data_adj_close,sp500_1_copy[i]["Adj Close"]], axis=1)
data_adj_close.columns=dfs_symbol
data_adj_close
```

```python
data_open=pd.DataFrame()
for i in dfs_symbol:
    data_open= pd.concat([data_open,sp500_1_copy[i]["Open"]], axis=1)
data_open.columns=dfs_symbol
data_open
```

```python
data_high=pd.DataFrame()
for i in dfs_symbol:
    data_high= pd.concat([data_high,sp500_1_copy[i]["High"]], axis=1)
data_high.columns=dfs_symbol
data_high
```

```python
data_low=pd.DataFrame()
for i in dfs_symbol:
    data_low= pd.concat([data_low,sp500_1_copy[i]["Low"]], axis=1)
data_low.columns=dfs_symbol
data_low
```

```python

data_close=pd.DataFrame()
for i in dfs_symbol:
    data_close= pd.concat([data_close,sp500_1_copy[i]["Close"]], axis=1)
data_close.columns=dfs_symbol
data_close
```

```python
data_volumen=pd.DataFrame()
for i in dfs_symbol:
    data_volumen= pd.concat([data_volumen,sp500_1_copy[i]["Volume"]], axis=1)
data_volumen.columns=dfs_symbol
data_volumen
```

## Realizo los calculos

Retorno de los movimientos GAP

```python
#formula
retorno_gap = np.log(data_open/data_close.shift(1)).fillna(0)
# creo una columna Dia
retorno_gap['Dia'] = retorno_gap.index.map(lambda x: x.weekday()+1)
#agrupo por los Dias sumandolos y calculando la media
retorno_gap_dia = pd.DataFrame(retorno_gap.groupby('Dia').sum().T.mean(), columns=['r_gap'])
retorno_gap_dia.index = ['Lunes','Martes','Miercoles','Jueves','Viernes']
```

```python
retorno_gap_dia
```

graficos movimientos gap.

##### El mejor dia para invetir teniendo en cuenta el retorno gap es el dia Martes

```python
fig = px.bar(retorno_gap_dia, x=retorno_gap_dia.index, y='r_gap', title='Retornos GAP')
fig.show()
```

##### Calculo de movimientos intradiarios

```python
#formula
retorno_intra = np.log(data_close/data_open).fillna(0)
# creo una columna Dia
retorno_intra['Dia'] = retorno_intra.index.map(lambda x: x.weekday()+1)
#agrupo por los Dias sumandolos y calculando la media
retorno_intra_dia = pd.DataFrame(retorno_intra.groupby('Dia').sum().T.mean(), columns=['r_intra'])
retorno_intra_dia.index = ['Lunes','Martes','Miercoles','Jueves','Viernes']
```

```python
retorno_intra_dia
```

Grafico de movimientos intradiarios.

El mejor dia para invitar teniendo en cuantos los movimientos intradiarios es el dia jueves

```python
fig1 = px.bar(retorno_intra_dia, x=retorno_intra_dia.index, y='r_intra', title='Retornos intradiarios')
fig1.show()
```

##### Calculo Variaciones

```python
#formula
variaciones = data_adj_close.pct_change()
# creo una columna Dia
variaciones['Dia'] = variaciones.index.map(lambda x: x.weekday()+1)
#agrupo por los Dias sumandolos y calculando la media
variaciones_dia = pd.DataFrame(variaciones.groupby('Dia').sum().T.mean(), columns=['variety'])
variaciones_dia.index = ['Lunes','Martes','Miercoles','Jueves','Viernes']
variaciones_dia
```

El dia con mayor variacion media es el dia Martes

```python
fig2 = px.bar(variaciones_dia, x=variaciones_dia.index, y='variety', title='Variety')
fig2.show()
```

##### Calculo volatilidad

```python
#formula
volatilidad = variaciones.rolling(250).std()*100*(250)**0.5
# creo una columna Dia
volatilidad['Dia'] = volatilidad.index.map(lambda x: x.weekday()+1)
#agrupo por los Dias sumandolos y calculando la media
volatilidad_dia = pd.DataFrame(volatilidad.groupby('Dia').sum().T.mean(), columns=['volatility'])
volatilidad_dia.index = ['Lunes','Martes','Miercoles','Jueves','Viernes']
```

```python
volatilidad_dia
```

grafico de volatilidad por dia.

El valor con mayor volatilidad es el dia Miercoles

```python
fig3 = px.bar(volatilidad_dia, x=volatilidad_dia.index, y='volatility', title='volatilidad por dia')
fig3.show()
```

### VIX

```python
#Calculo el vix aplicando la transpuesta para que quede por fechas
Vix_mayor = volatilidad.T.mean().sort_values(ascending=False)
Vix_mayor
```

##### Como se puede apreciar, la fecha de mas alta volatilidad que afecto al sp500 fue en junio del 2009, que empezo con un alto valor la primera de semana y empezo a descender al final de la semana del mismo mes

## Grafico de volitdad entre los años 2020-2021

```python
Vix=volatilidad.T.mean()
fig = px.line(x=Vix.index, y=Vix, width=1500, height=400, title="Grafico de Volatilidad entre los años 2000-2021 ")
fig.show()
```

##### variacion historica de las mejores empresas del sp500 desde el año 2000 hasta el 2021 inclusive en las que invertir 

```python
data_change_percent= (((data_close.fillna(method='bfill').iloc[-1:])-(data_open.fillna(method='bfill').iloc[0,:]))/(data_open.fillna(method='bfill').iloc[0,:]))*100
data_change_percent=pd.DataFrame(data_change_percent.sum(), columns=["Variacion Total"])
data_change_percent=data_change_percent.sort_values(by="Variacion Total", ascending=False)
data_change_percent[:10]
```

```python
fig3 = px.bar(data_change_percent[:10], x=data_change_percent[:10].index, y='Variacion Total', title='top 10 mejores empresas para invertir')
fig3.show()
```

```python
data_change_percent.reset_index(inplace=True)
```

```python
data_change_percent.rename(columns={'index':'Symbol'}, inplace=True)
```

```python

data = pd.read_html('https://en.wikipedia.org/wiki/List_of_S%26P_500_companies')[0][['Symbol','GICS Sector','GICS Sub-Industry']].set_index(['Symbol'])
```

```python
data_change_percent = data_change_percent.merge(data, how='left', on='Symbol')
```

```python
data_change_percent
```

#### mejores sectores del sp500 entre el año 2000 y el año 2021

```python
top_sector = data_change_percent.groupby('GICS Sector').sum().sort_values(by=['Variacion Total'], ascending=False)
top_sector
```

```python
fig4 = px.bar(top_sector, x=top_sector.index, y='Variacion Total', title='Sectores del sp500 mejores posicionados')
fig4.show()
```

#### Mejores industrias del sp500 entre los años 2000 y 2021

```python
top_industria = data_change_percent.groupby('GICS Sub-Industry').sum().sort_values(by=['Variacion Total'], ascending=False)[:10]
top_industria
```

```python
fig5 = px.bar(top_industria, x=top_industria.index, y='Variacion Total', title='Industrias del sp500 mejores posicionados')
fig5.show()
```
